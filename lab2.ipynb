```python
#1c
reset()
x=var('x')
y=function('y')(x)
ec1=diff(y,x)-2*x(1+y^2)==0
e=desolve(ec1,y)
[sol]=solve(e,y)
f(x,_C)=sol.rhs()
f(x,_C)
```
