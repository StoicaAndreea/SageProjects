```python
#1a
reset()
t,k,x0=var('t,k,x0')
x=function('x')(t)
ec1=diff(x(t),t)==-k*x
f(t,k,x0)=desolve(ec1,[x,t],[0,x0])
f(t,k,x0)
```

```python
#1b
plot([f(t,1/8000,2),f(t,1/8000,4),f(t,1/8000,3)],t,0,100000,ymin=0,ymax=5)
```

```python
T12=var('T12')
ec2=f(T12,k,x0)==x0/2
[k1]=solve(ec2,k)
k2(T12)=k1.rhs()
k2(T12)
```

```python
numerical_approx(k2(5730))
```

```python
#3ab malthus
reset()
t,k,x0=var('t,k,x0')
x=function('x')(t)
ec1=diff(x(t),t)==k*x
f(t,k,x0)=desolve(ec1,[x,t],[0,x0])
f(t,k,x0)
```

```python
plot([f(t,0.1,10),f(t,0.3,10),f(t,0.5,2)],t,0,30,ymin=0,ymax=300)
```

```python
#3ab Verhulst
reset()
t,k,x0,G=var('t,k,x0,G')
x=function('x')(t)
ec1=diff(x(t),t)==k*x*(G-x)
sol=desolve(ec1,[x,t],[0,x0])
[sol1]=solve(sol,x)
sol1
```

```python
sol1.log_simplify()
f(t,k,x0,G)=exp(G*k*t - log(-G + x0) + log(-G + x(t)) + log(x0))
f(t,k,x0,G)
```

```python
plot([f(t,10,200,0.1),f(t,10,200,0.2),f(t,10,300,0.3)],t,0,58,ymin=0,ymax=300)
```
